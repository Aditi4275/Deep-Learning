```python
import tensorflow
from tensorflow import keras
from keras.preprocessing import image
from tensorflow.keras.preprocessing.image import ImageDataGenerator
```

```python
img = image.load_img('/content/cat.jpg', target_size=(200,200))
```

```python
img
```

```python
type(img)
```

```python
datagen = ImageDataGenerator(
    rotation_range = 30,     #amout of rotation
    shear_range = 0.2,       #distortion
    zoom_range = 0.2,        #zoom in or zoom out
    horizontal_flip = True,  #right to left & vv
    width_shift_range = 0.2, #shifting the width
    height_shift_range = 0.2,#shifting the height
    fill_mode = 'reflect'    #mode to fill the space when image is shifted
                             # reflect- reflects the data, constant - leave the black space, nearest - fill by nearest traits
)
```

```python
img = image.img_to_array(img)
```

```python
type(img)
```

```python
img.shape
```

```python
input_batch = img.reshape(1,200,200,3)    # 1 since only 1 image is in batch
```

```python
import os
if not os.path.exists('preview'):
    os.makedirs('preview')
i = 0
for output in datagen.flow(input_batch,batch_size=1,save_to_dir='preview',save_prefix='cat',save_format='jpg'):
  i += 1
  if i == 10:
    break
```
